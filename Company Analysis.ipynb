```python
uniq_vals = df1['Design'].unique()
```

```python
skillscore_of_profession = []
for i in uniq_vals:
    newdf = df1[df1["Design"]==i]
    
    skillscore_of_profession.append([i,newdf["Skill Score"].mean()])
sorted_skillscore_of_profession = sorted(skillscore_of_profession , key = lambda x:x[1])
```

```python
uniq_vals_company = df1['Company'].unique()
```

```python
skillscore_of_company = []
for i in uniq_vals_company:
    newdf = df1[df1["Company"]==i]
    
    skillscore_of_company.append([i , newdf["Skill Score"].mean()])
sorted_company_by_skill = sorted(skillscore_of_company,key = lambda  x:x[1],reverse = True)
```

```python
#Top 10 companies with highest Average skill score
k=1
for i in sorted_company_by_skill[:10]:
    print(str(k)+"."+str(i[0])+"                "+"  : "+str(i[1]))
    k+=1
```

```python
# Top 10 proffessions with highest Mean Skill Score
k=1

for i in sorted_skillscore_of_profession[:10]:
    print(str(k)+"."+str(i[0])+"  "+"  : "+str(i[1]))
    k+=1
```
